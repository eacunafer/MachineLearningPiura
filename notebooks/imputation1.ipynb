```python
#cotejando si la tabla tiene missing values
df.isnull()
```

```python
#Hallando el numero de entradas de la tabla que son missing
df.isnull().sum().sum()
```

```python
colmiss=df.columns[df.isnull().any()].tolist()
print colmiss
```

```python
#Hallando el numero de entradas missing de cada columna
df[colmiss].isnull().sum()
```

```python
#Otra manera de hallar el numero de entradas missings en cada columna
df.apply(lambda x: sum(x.isnull().values), axis = 0)
```

```python
#Hallando el numero de filas con missing
df.isnull().T.any().T.sum()
```

```python
#Otra manera: Hallando el numero de filas que tienen missings
sum(df.apply(lambda x: sum(x.isnull().values), axis = 1)>0) 
```

```python
#Una tercera manera de hallar el numero de filas con missings
sum([True for idx,row in df.iterrows() if any(row.isnull())])
```

```python
#Eliminando  las filas que tienen missing
dfclean=df.dropna() 
```
